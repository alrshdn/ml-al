```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://gist.githubusercontent.com/meperezcuello/82a9f1c1c473d6585e750ad2e3c05a41/raw/d42d226d0dd64e7f5395a0eec1b9190a10edbc03/Medical_Cost.csv")
```

```python
data.head()
```

```python
# data[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
for i in range(1338):
  if data['region'][i] == "northeast":
    data['region'][i] = 1.0
  elif data['region'][i] == "southeast":
    data['region'][i] = 2.0
  elif data['region'][i] == "southwest":
    data['region'][i] = 3.0
  elif data['region'][i] == "northwest":
    data['region'][i] = 4.0

  if data['sex'][i] == 'male':
    data['sex'][i] = 0.0
  else:
    data['sex'][i] = 1.0

  if data['smoker'][i] == 'no':
    data['smoker'][i] = 0.0
  else:
    data['smoker'][i] = 1.0
```

```python
X = data.drop(columns='charges')
y = data['charges']
X_train, y_train, X_test, y_test = X.iloc[0:1070], y.iloc[0:1070], X.iloc[1070:], y.iloc[1070:]
X_train = np.asarray(X_train).astype('float32')
y_train = np.asarray(X_train).astype('float32')
X_test = np.asarray(X_test).astype('float32')
y_test = np.asarray(y_test).astype('float32')

X_train = tf.expand_dims(tf.constant(X_train, dtype=tf.float64), axis=0)
y_train = tf.expand_dims(tf.constant(y_train, dtype=tf.float64), axis=0)
X_test = tf.expand_dims(tf.constant(X_test, dtype=tf.float64), axis=0)
y_test = tf.expand_dims(tf.constant(y_test, dtype=tf.float64), axis=0)
y_test.shape

y_test.shape == X_test.shape
X_train.size
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(6)
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(lr=0.01),
    loss=tf.keras.losses.mae,
    metrics=["mae"])

model.fit(X_train, y_train, epochs=1000, verbose=0)
```

```python
model.evaluate(X_test, y_test)
# np.array(y_test).max() - np.array(y_test).min()

#y_test
#pd.DataFrame.median()
```
